1. Given an array arr[ ] of size N having elements, the task is to find the next greater element for each element of the array in order of their appearance in the array.Next greater element of an element in the array is the nearest element on the right which is greater than the current element.If there does not exist next greater of current element, then next greater element for current element is -1. For example, next greater of the last element is always -1.

Example 1:
Input:
N = 4, arr[] = [1 3 2 4]
Output:
3 4 4 -1

Example 2:
Input:
N = 5, arr[] [6 8 0 1 3]
Output:
8 -1 1 3 -1


```python
def next_greater_element(arr):
    n = len(arr)
    stack = []
    result = [-1] * n

    for i in range(n - 1, -1, -1):
        while stack and arr[i] > stack[-1]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(arr[i])

    return result


arr = [1, 3, 2, 4]
print("arr[] : ",arr )
result = next_greater_element(arr)
print("Output : ",result)

print("\n")

arr = [6, 8, 0, 1, 3]
print("arr[] : ",arr )
result = next_greater_element(arr)
print("Output : ",result)
```

2. Given an array a of integers of length n, find the nearest smaller number for every element such that the smaller element is on left side.If no small element present on the left print -1.

Example 1:
Input: n = 3
a = {1, 6, 2}
Output: -1 1 1

Example 2:
Input: n = 6
a = {1, 5, 0, 3, 4, 5}
Output: -1 1 -1 0 3 4

```python
def nearest_smaller_number(arr):
    n = len(arr)
    stack = []
    result = [-1] * n

    for i in range(n):
        while stack and arr[i] <= stack[-1]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(arr[i])

    return result

# Example taken from the question

arr = [1, 6, 2]
print("arr[] : ",arr )
result = nearest_smaller_number(arr)
print("Output : ",result)

print("\n")

arr = [1, 5, 0, 3, 4, 5]
print("arr[] : ",arr )
result = nearest_smaller_number(arr)
print("Output : ",result)
```

3. Implement a Stack using two queues q1 and q2.

Example 1:
Input:
push(2)
push(3)
pop()
push(4)
pop()
Output:  3 4


Example 2:
Input:
push(2)
pop()
pop()
push(3)
Output:   2 -1

```python
class Stack:
    def __init__(self):
        self.q1 = []
        self.q2 = []

    def push(self, x):
        self.q1.append(x)

    def pop(self):
        if not self.q1 and not self.q2:
            return -1

        if not self.q2:
            while len(self.q1) > 1:
                self.q2.append(self.q1.pop(0))

        return self.q1.pop(0)

    def top(self):
        if not self.q1 and not self.q2:
            return -1

        if not self.q2:
            while len(self.q1) > 1:
                self.q2.append(self.q1.pop(0))

        return self.q1[0]

    def isEmpty(self):
        return not self.q1 and not self.q2

# Example taken from the question

print("Output: ")
stack = Stack()
stack.push(2)
stack.push(3)
print(stack.pop())  
stack.push(4)
print(stack.pop())  

print("\n")

print("Output: ")
stack = Stack()
stack.push(2)
print(stack.pop()) 
print(stack.pop())  
stack.push(3)
print(stack.top())
```

4. You are given a stack St. You have to reverse the stack using recursion.

Example 1:
Input:St = {3,2,1,7,6}
Output:{6,7,1,2,3}

Example 2:
Input:St = {4,3,9,6}
Output:{6,9,3,4}

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.items.pop()

    def peek(self):
        if not self.isEmpty():
            return self.items[-1]

    def isEmpty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


def insert_bottom(stack, item):
    if stack.isEmpty():
        stack.push(item)
    else:
        temp = stack.pop()
        insert_bottom(stack, item)
        stack.push(temp)


def reverseStack(stack):
    if not stack.isEmpty():
        temp = stack.pop()
        reverseStack(stack)
        insert_bottom(stack, temp)


def print_stack(stack):
    print("Original Stack:")
    for item in stack.items:
        print(item, end=" ")
    print()
        
# Example taken from the question

stack = Stack()
stack.push(3)
stack.push(2)
stack.push(1)
stack.push(7)
stack.push(6)
print_stack(stack)
print("Output: ")
reverseStack(stack)
print(stack.items) 

print("\n")

stack = Stack()
stack.push(4)
stack.push(3)
stack.push(9)
stack.push(6)
print_stack(stack)
print("Output: ")
reverseStack(stack)
print(stack.items) 
```

5. You are given a string S, the task is to reverse the string using stack.

Example 1:
Input: S="GeeksforGeeks"
Output: skeeGrofskeeG

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.items.pop()

    def peek(self):
        if not self.isEmpty():
            return self.items[-1]

    def isEmpty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


def reverseString(s):
    stack = Stack()
    for char in s:
        stack.push(char)

    reversed_str = ''
    while not stack.isEmpty():
        reversed_str += stack.pop()

    return reversed_str


# Example taken from the question

S = input("Input : ")
reversed_S = reverseString(S)
print("Reversed String:", reversed_S)
```

6. Given string S representing a postfix expression, the task is to evaluate the expression and find the final value. Operators will only include the basic arithmetic operators like , /, + and -.

Example 1:
Input: S = "231*+9-"
Output: -4

Example 2:
Input: S = "123+*8-"
Output: -3


```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.items.pop()

    def peek(self):
        if not self.isEmpty():
            return self.items[-1]

    def isEmpty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


def evaluatePostfix(S):
    stack = Stack()
    operators = set(['+', '-', '*', '/'])

    for char in S:
        if char not in operators:
          
            stack.push(int(char))
        else:
          
            operand2 = stack.pop()
            operand1 = stack.pop()
          
            if char == '+':
                result = operand1 + operand2
            elif char == '-':
                result = operand1 - operand2
            elif char == '*':
                result = operand1 * operand2
            elif char == '/':
                result = operand1 / operand2
           
            stack.push(result)

    return stack.pop()


# Example taken from the question

S = "231*+9-"
print("Input: ",S)
result = evaluatePostfix(S)
print("Result:", result)

print("\n")

S = "123+*8-"
print("Input: ",S)
result = evaluatePostfix(S)
print("Result:", result)
```

7. Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the `MinStack` class:

- `MinStack()` initializes the stack object.
- `void push(int val)` pushes the element `val` onto the stack.
- `void pop()` removes the element on the top of the stack.
- `int top()` gets the top element of the stack.
- `int getMin()` retrieves the minimum element in the stack.

You must implement a solution with `O(1)` time complexity for each function.

Example 1:
Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]


```python
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []  

    def push(self, val):
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self):
        if self.stack:
            val = self.stack.pop()
            if val == self.min_stack[-1]:
                self.min_stack.pop()

    def top(self):
        if self.stack:
            return self.stack[-1]

    def getMin(self):
        if self.min_stack:
            return self.min_stack[-1]
        else:
            return "Null"


# Example taken from the question

stack = MinStack()
stack.push(-2)
stack.push(0)
stack.push(-3)
output1 = stack.getMin()  
stack.pop()
output2 = stack.top() 
output3 = stack.getMin()  

output = [None, None, None, None, output1, None, output2, output3]
print(output)
```
