1. Given a binary tree, your task is to find subtree with maximum sum in tree.

Example 1:

Input :       

       1

     /   \

   2      3

  / \    / \

4   5  6   7

Output : 28



Example 2:

Input :

       1

     /    \

  -2      3

  / \    /  \

4   5  -6   2

Output : 7

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def find_sub_tree_sum(node):
    if node is None:
        return 0

    left_sum = find_sub_tree_sum(node.left)
    right_sum = find_sub_tree_sum(node.right)
    subtree_sum = node.val + left_sum + right_sum

    return subtree_sum


def find_max_sub_tree_sum(root):
    if root is None:
        return 0

    max_sum = float('-inf')

    def dfs(node):
        nonlocal max_sum
        if node is None:
            return 0

        left_sum = dfs(node.left)
        right_sum = dfs(node.right)
        subtree_sum = node.val + left_sum + right_sum

        max_sum = max(max_sum, subtree_sum)

        return subtree_sum

    dfs(root)

    return max_sum



# Example taken from the question

root1 = TreeNode(1)
root1.left = TreeNode(2)
root1.right = TreeNode(3)
root1.left.left = TreeNode(4)
root1.left.right = TreeNode(5)
root1.right.left = TreeNode(6)
root1.right.right = TreeNode(7)
print(find_max_sub_tree_sum(root1))


root2 = TreeNode(1)
root2.left = TreeNode(-2)
root2.right = TreeNode(3)
root2.left.left = TreeNode(4)
root2.left.right = TreeNode(5)
root2.right.left = TreeNode(-6)
root2.right.right = TreeNode(2)
print(find_max_sub_tree_sum(root2))
```

2. Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10

```python
from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def constructBst(arr, n):
    queue = deque()
    if n == 0:
        return None

    root = Node(arr[0])
    head = root

    queue.append((root, (-float("inf"), float("inf"))))
    i = 1

    while i < n:
        temp = queue[0][0]
        tempRange = queue[0][1]
        queue.popleft()

        if (arr[i] < temp.data) and tempRange[0] < arr[i] < tempRange[1]:
            temp.left = Node(arr[i])
            queue.append((temp.left, (tempRange[0], temp.data)))
            i += 1


        if arr[i] > temp.data and tempRange[0] < arr[i] < tempRange[1]:
            temp.right = Node(arr[i])
            queue.append((temp.right, (temp.data, tempRange[1])))
            i += 1
    return head


def inorderTraversal(root):
    if (root == None):
        return None

    inorderTraversal(root.left)
    print(root.data, end=" ")
    inorderTraversal(root.right)


if __name__ == '__main__':

    arr = [7, 4, 12, 3, 6, 8, 1, 5, 10]
    n = len(arr)
    root = constructBst(arr, n)

    print("Inorder Traversal: ")
    root = inorderTraversal(root)
```

3. Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Example 1:

Input : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}
Output : Yes

Example 2:

Input : arr[] = {11, 6, 13, 5, 12, 10}
Output : No

```python
def canRepresentBST(arr):
    if len(arr) == 0:
        return "No"

    n = len(arr)
    stack = []
    root = float("-inf")
    for i in range(n):
        if arr[i] < root:
            return "No"

        while len(stack) > 0 and arr[i] > stack[-1]:
            root = stack.pop()

        stack.append(arr[i])

    return "Yes"

# Example taken from the question

arr2 = [11, 6, 13, 5, 12, 10]
print(canRepresentBST(arr2)) 
```
